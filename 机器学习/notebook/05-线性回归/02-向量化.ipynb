```python
import numpy as np
import matplotlib.pyplot as plt

# 样本数据
x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])

# 绘图
plt.scatter(x, y)
plt.axis([0, 6, 0, 6])
plt.show()
```

# 求a、b值

# 向量化

```python
from playML.SimpleLinearRegression import SimpleLinearRegression2

reg1 = SimpleLinearRegression2()
reg1.fit(x, y)

x_predict = 6
# 输入6，得到预测值
y_predict = reg1.predict(np.array([x_predict]))
print(y_predict)

#  获取a、b的值
print(reg1.a_)
print(reg1.b_)
```

```python
y_hat1 = reg1.a_*x + reg1.b_
plt.scatter(x, y)
plt.plot(x, y_hat1, color='r')
plt.axis([0, 6, 0, 6])
plt.show()
```
