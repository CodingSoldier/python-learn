```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
#  绘制x、y坐标轴图形
plt.plot(x, y)
plt.show()
```

```python
# 绘制sin、cos两条曲线图像
cosy = np.cos(x)
siny = y.copy()
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
# 指定x、y轴图示说明
plt.plot(x, siny, label="sin(x)")
# 指定颜色，线条
plt.plot(x, cosy, color='red', linestyle="--", label="cos(x)")
# 修改X坐标轴
plt.xlim(-5, 15)
# 修改Y轴
plt.ylim(0, 1.5)
# x、y标签 
plt.xlabel("x axis")
plt.ylabel("y value")
# 显示图示
plt.legend()
# 标题
plt.title("welcome to xxx")
plt.show()
```

```python
# 散点图
plt.scatter(x, siny)
plt.show()
```

```python
# 两条散点图
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```

```python
# 散点图，二维正态分布
x = np.random.normal(0, 1, 1000)
y= np.random.normal(0, 1, 1000)
plt.scatter(x, y, alpha=0.5)
plt.show()
```
